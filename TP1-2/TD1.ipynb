```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as pt
pt.rcParams.update({"text.usetex": False, "font.family": "Times"})
import control as ct
from control.matlab import *   
```

```python
A = np.array([-.2])
B = np.array([[.2]])
C = np.array([[5]])
D = np.array([[0.0]])
```

```python
h1 = 2
h2 = 2
a1 = 0.01
a2 = 0.01
b1 = 0.005
b2 = 0.005
```

```python
ks = 5
kp = 1
y0 = 20
x0 = np.array([1/ks * y0])
t0, tf = 0,30
t = np.linspace(t0, tf, 100)
```

```python
HeatedRoom = ss(A, B, C, D)
Thermostat = ct.tf(kp,1)
System = series(Thermostat, HeatedRoom)
closed_loop = feedback(System, 1)
```

```python
#y,t = impulse(HeatedRoom)
#y,t = step(HeatedRoom)
y,t = step(System)
```

```python
print(y.shape)
print(t.shape)
```

```python
pt.plot(t, y, linewidth=2.0, color='blue')
pt.grid(color='lightgray', linestyle='--')
pt.xlim([t0, tf])
pt.xlabel('Time (s)')
pt.ylabel(r'Temperature (\textdegree C)')
pt.title('Room heating')
pt.show()
```

```python

h1 = 2.0
h2 = 2.0
a1 = 0.01
a2 = 0.01
b1 = 0.005
b2 = 0.005

A = np.array([[-a1,  b1],
              [ b2, -a2]])

B = np.array([[h1],
              [0.0]])

C = np.eye(2)
D = np.zeros((2, 1))

sys_l1 = ss(A, B, C, D)
print(sys_l1)
```

```python
t_end = 100.0
t = np.linspace(0, t_end, 1001)

t_out, y_out = ct.step_response(sys_l1, T=t)
print(y_out.shape)
print(t_out.shape)
```

```python

x1 = np.squeeze(y_out[0, :])
x2 = np.squeeze(y_out[1, :])

print(x1.shape)
print(x2.shape)
```

```python

pt.figure()
pt.plot(t_out, x1, label=r'$x_1(t)$ (tank 1)')
pt.plot(t_out, x2, label=r'$x_2(t)$ (tank 2)')
pt.xlabel('time')
pt.ylabel('temperature')
pt.title('Step response of mode $l_1$ (ON$_1$  OFF$_2$)')
pt.grid(True)
pt.legend()
pt.show()
```

```python
x0 = np.array([0.0, 50.0])

u = np.ones_like(t)

t_forced, y_forced = ct.forced_response(sys_l1, T=t, U=u, X0=x0)

x1_sim = np.squeeze(y_forced[0, :])
x2_sim = np.squeeze(y_forced[1, :])

pt.figure()
pt.plot(t_forced, x1_sim, label=r'$x_1(t)$')
pt.plot(t_forced, x2_sim, label=r'$x_2(t)$')
pt.xlabel('time')
pt.ylabel('temperature (arbitrary units)')
pt.title(r'Simulation of mode $l_1$ with $x_0 = (0, 50)^T$')
pt.grid(True)
pt.legend()
pt.show()
```
